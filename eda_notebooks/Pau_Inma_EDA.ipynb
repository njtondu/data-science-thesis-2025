## 4. Análisis profundo

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculamos la media (o puedes usar df.sum() si prefieres la suma total)
entries = [
    'Entries_Mon', 'Entries_Midweek', 'Entries_Fri', 'Entries_Sat', 'Entries_Sun'
]
exits = [
    'Exits_Mon', 'Exits_Midweek', 'Exits_Fri', 'Exits_Sat', 'Exits_Sun'
]

# Extraemos la media por día
entries_mean = df[entries].mean()
exits_mean = df[exits].mean()

# Extraemos solo el día para etiquetas (quitar "Entries_" y "Exits_")
dias = ['Mon', 'Midweek', 'Fri', 'Sat', 'Sun']
```

```python
import numpy as np

x = np.arange(len(dias))  # posiciones de los grupos
width = 0.35  # ancho de las barras

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, entries_mean.values, width, label='Entries', color='skyblue')
plt.bar(x + width/2, exits_mean.values, width, label='Exits', color='salmon')

# Etiquetas y diseño
plt.xlabel('Día de la semana')
plt.ylabel('Promedio de personas')
plt.title('Entradas vs Salidas por día de la semana')
plt.xticks(x, dias)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# cORRELACIONES
import seaborn as sns

corr_matrix = df[[*entry_cols, *exit_cols]].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlación entre entradas y salidas por día")
plt.show()
```

```python
# Top 10 estaciones por tráfico semanal
top_stations = df[['Station', 'Weekly_EnEx']].sort_values(by='Weekly_EnEx', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.barh(top_stations['Station'], top_stations['Weekly_EnEx'], color='mediumseagreen')
plt.title("Top 10 estaciones por tráfico semanal")
plt.xlabel("Weekly_EnEx")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# VEMOS SI HAY ESTACIONES DESEQUILIBRADAS
# Asegúrate de tener las columnas limpias
df.columns = df.columns.str.strip()

# Lista de días
days = ['Mon', 'Midweek', 'Fri', 'Sat', 'Sun']
entry_cols = [f'Entries_{day}' for day in days]
exit_cols = [f'Exits_{day}' for day in days]

# Sumar entradas y salidas semanales por estación
df['Total_Entries_Week'] = df[entry_cols].sum(axis=1)
df['Total_Exits_Week'] = df[exit_cols].sum(axis=1)

# Calcular diferencia (positiva = más entradas; negativa = más salidas)
df['EntryExit_Diff'] = df['Total_Entries_Week'] - df['Total_Exits_Week']

# Estación con más diferencia positiva (entra más gente de la que sale)
max_positive = df.loc[df['EntryExit_Diff'].idxmax(), ['Station', 'EntryExit_Diff']]

# Estación con más diferencia negativa (sale más gente de la que entra)
max_negative = df.loc[df['EntryExit_Diff'].idxmin(), ['Station', 'EntryExit_Diff']]

print("📈 Estación con más entradas que salidas:")
print(max_positive)

print("\n📉 Estación con más salidas que entradas:")
print(max_negative)
```
